```python
import numpy as np

def jacobi(A, b):
    n = len(b)
    x = np.zeros(n)  

    print("Método de Jacobi")
    print("x(0) =", x)

    for k in range(1, 3):  
        x_new = np.zeros(n)

        for i in range(n):
            suma = 0
            for j in range(n):
                if j != i:
                    suma += A[i][j] * x[j]
            x_new[i] = (b[i] - suma) / A[i][i]

        print(f"x({k}) =", x_new)
        x = x_new.copy()

    return x
```
