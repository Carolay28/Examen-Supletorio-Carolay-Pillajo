```python
import numpy as np
import matplotlib.pyplot as plt


def jacobi(A, b):
    n = len(b)
    x = np.zeros(n)  

    print("MÃ©todo de Jacobi")
    print("x(0) =", x)

    for k in range(1, 3):  
        x_new = np.zeros(n)

        for i in range(n):
            suma = 0
            for j in range(n):
                if j != i:
                    suma += A[i][j] * x[j]
            x_new[i] = (b[i] - suma) / A[i][i]

        print(f"x({k}) =", x_new)
        x = x_new.copy()

    return x
```

```python
A = np.array([[2,1],[3,2]], dtype=float)
b = np.array([5,8], dtype=float)

tray = []
n = len(b)
x = np.zeros(n)
tray.append(x.copy())

for k in range(1, 8):
    x_new = np.zeros(n)
    for i in range(n):
        suma = 0
        for j in range(n):
            if j != i:
                suma += A[i][j] * x[j]
        x_new[i] = (b[i] - suma) / A[i][i]
    tray.append(x_new.copy())
    x = x_new.copy()

tray = np.array(tray)

jacobi(A, b)

plt.plot(tray[:,0], tray[:,1], "o:", markersize=4)
plt.plot(tray[0,0], tray[0,1], "gd", markersize=8, label="$x_0$")
plt.plot(tray[-1,0], tray[-1,1], "ro", markersize=8, label="$x_s$")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Trayectoria Jacobi - Ejercicio 1")
plt.legend()
plt.grid(True)
plt.show()
```
