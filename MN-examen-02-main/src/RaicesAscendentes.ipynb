- Usando los métodos estudiados, determine todas las raíces de la ecuación:

$x^5+5.25*x^4+4.125*x^3−9.125*x^2−14.625*x−5.625=0$

```python
import numpy as np
from scipy.optimize import bisect, newton
```

```python
def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625
```

```python
def df(x):
    return 5*x**4 + 21*x**3 + 12.375*x**2 - 18.25*x - 14.625
```

```python
xs = np.linspace(-6, 6, 2000)
intervalos = []

for i in range(len(xs)-1):
    if f(xs[i]) * f(xs[i+1]) < 0:
        intervalos.append((xs[i], xs[i+1]))

intervalos
```

```python
roots_bisect = []

for a, b in intervalos:
    r = bisect(f, a, b)
    roots_bisect.append(r)

roots_bisect
```

```python
roots_newton = []

for r in roots_bisect:
    rn = newton(f, r, fprime=df)
    roots_newton.append(rn)

roots_newton
```

```python
roots = np.unique(np.round(roots_newton, 6))
roots = np.sort(roots)
roots
```

```python
while len(roots) < 6:
    roots = np.append(roots, np.nan)

roots
```
