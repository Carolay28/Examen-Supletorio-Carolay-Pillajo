```python
import numpy as np
import matplotlib.pyplot as plt
```

EJERCICIO 1

A = [[2,1],[3,2]]
b = [5,8]

```python
def jacobi(A, b, x0, it):
    D = np.diag(np.diag(A))
    R = A - D
    D_inv = np.linalg.inv(D)
    T = -D_inv @ R
    c = D_inv @ b
    
    x = x0.copy()
    tray = [x.copy()]
    
    for k in range(it):
        x = T @ x + c
        tray.append(x.copy())
    
    return np.array(tray), T
```

```python
A1 = np.array([[2,1],
               [3,2]], dtype=float)

b1 = np.array([5,8], dtype=float)

x0 = np.array([0,0], dtype=float)

tray1, T1 = jacobi(A1, b1, x0, 15)
```

```python
sol1 = np.linalg.solve(A1, b1)
sol1
```

```python
vals1, vecs1 = np.linalg.eig(T1)

idx1 = np.argmax(np.abs(vals1))

v1 = vecs1[:, idx1]

m1 = v1[1] / v1[0]

m1
```

```python
plt.plot(tray1[:,0], tray1[:,1], 'o-')
plt.scatter(sol1[0], sol1[1])
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```
