```python
def def_Grado4_parcial_4(xs: list, ys: list) -> tuple[float, float, float, float, float, float]:
    c_4 = sum(xi**8 for xi in xs)
    c_3 = sum(xi**7 for xi in xs)
    c_2 = sum(xi**6 for xi in xs)
    c_1 = sum(xi**5 for xi in xs)
    c_0 = sum(xi**4 for xi in xs)
    c_ind = sum(ys[i] * xi**4 for i, xi in enumerate(xs))
    return (c_4, c_3, c_2, c_1, c_0, c_ind)


def def_Grado4_parcial_3(xs: list, ys: list) -> tuple[float, float, float, float, float, float]:
    c_4 = sum(xi**7 for xi in xs)
    c_3 = sum(xi**6 for xi in xs)
    c_2 = sum(xi**5 for xi in xs)
    c_1 = sum(xi**4 for xi in xs)
    c_0 = sum(xi**3 for xi in xs)
    c_ind = sum(ys[i] * xi**3 for i, xi in enumerate(xs))
    return (c_4, c_3, c_2, c_1, c_0, c_ind)


def def_Grado4_parcial_2(xs: list, ys: list) -> tuple[float, float, float, float, float, float]:
    c_4 = sum(xi**6 for xi in xs)
    c_3 = sum(xi**5 for xi in xs)
    c_2 = sum(xi**4 for xi in xs)
    c_1 = sum(xi**3 for xi in xs)
    c_0 = sum(xi**2 for xi in xs)
    c_ind = sum(ys[i] * xi**2 for i, xi in enumerate(xs))
    return (c_4, c_3, c_2, c_1, c_0, c_ind)


def def_Grado4_parcial_1(xs: list, ys: list) -> tuple[float, float, float, float, float, float]:
    c_4 = sum(xi**5 for xi in xs)
    c_3 = sum(xi**4 for xi in xs)
    c_2 = sum(xi**3 for xi in xs)
    c_1 = sum(xi**2 for xi in xs)
    c_0 = sum(xi for xi in xs)
    c_ind = sum(ys[i] * xi for i, xi in enumerate(xs))
    return (c_4, c_3, c_2, c_1, c_0, c_ind)


def def_Grado4_parcial_0(xs: list, ys: list) -> tuple[float, float, float, float, float, float]:
    c_4 = sum(xi**4 for xi in xs)
    c_3 = sum(xi**3 for xi in xs)
    c_2 = sum(xi**2 for xi in xs)
    c_1 = sum(xi for xi in xs)
    c_0 = len(xs)
    c_ind = sum(ys)
    return (c_4, c_3, c_2, c_1, c_0, c_ind)
```
