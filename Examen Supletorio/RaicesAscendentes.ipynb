```python
import numpy as np
from scipy import optimize
```

```python
def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625
```

```python
coef = [1, 5.25, 4.125, -9.125, -14.625, -5.625]

all_roots = np.roots(coef)
```

```python
real_roots = []

for r in all_roots:
    if np.abs(r.imag) < 1e-6:
        real_roots.append(r.real)

real_roots = list(np.sort(real_roots))

while len(real_roots) < 6:
    real_roots.append(np.nan)

print(real_roots)
```
